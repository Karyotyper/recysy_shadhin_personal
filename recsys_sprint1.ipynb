```python
def song_recommendation(query_index, model_knn, pivot_table, nrec):
    import numpy as np
    res = []
    user_behaviour = pivot_table.iloc[query_index, :].values.reshape(1, -1)
    distances, indices = model_knn.kneighbors(user_behaviour, n_neighbors = nrec+1)
    for i in range(0, len(distances.flatten())):
        if i == 0:
            continue
        else:
            res.append(pivot_table.index[indices.flatten()[i]])
    
    return res
```

```python
def load(filename):
    """
    Load from filename using pickle

    @param filename: name of file to load from
    @type filename: str
    """

    try:
        f = bz2.BZ2File(filename, 'rb')
    except IOError as details:
        sys.stderr.write('File ' + filename + ' cannot be read\n')
        sys.stderr.write(details)
        return

    myobj = pickle.load(f)
    f.close()
    return myobj
```
